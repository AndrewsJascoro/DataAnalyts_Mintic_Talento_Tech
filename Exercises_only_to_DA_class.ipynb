**The technical term for incorrect, inconsistent, or meaningless data in a dataset is "dirty data."**

**Note:**

**Data entry errors:** Typos, misspellings, and incorrect values entered by humans.

**Inconsistent data:** Inconsistent formatting, units, or representations of the same data across different records or columns.

**Missing data:** Empty or null values where data should be present.

**Duplicate data:** Duplicate records or entries within the dataset.

**Outliers:** Data points that are significantly different from other data points in the dataset, potentially indicating errors or anomalies.

**Abbreviations and acronyms:** Inconsistent or ambiguous abbreviations or acronyms used within the data.

**Inaccurate data:** Data that is outdated, incorrect, or no longer relevant.

**Non-standard values:** Values that do not conform to the expected range or format for a particular data field.

**Duplicate data:** Data that is poorly written, unclear, or unintelligible.

**Inconsistent capitalization:** Inconsistent use of uppercase and lowercase letters in text data.

```python
# Made by Javier Coronado for the Talento Tech data analysis and visualization bootcamp, Mintic, Bogota, Col.
import pandas as pd
import matplotlib.pyplot as pl
from IPython.display import display
import seaborn as sns

# 0. Define the path for the Parquet file
parquet_file_path = "dataset_banco_1_2.parquet" #"dataset_banco_101_cleaned_copy.parquet"
# Read the Parquet file into a pandas DataFrame
df = pd.read_parquet(parquet_file_path)

# 1. To find all the unique values in a column of a DataFrame
unique_education_status = df['marital'].unique()
print('Status or civil marital with dirty data:\n',unique_education_status)

# 2. Replace 'div.' and 'DIVORCED' with 'divorced' in the 'marital' column
df['marital'] = df['marital'].replace(['div.', 'DIVORCED'], 'divorced')
df['marital'] = df['marital'].replace(['Single'], 'single')

# 3. unique matital data
unique_marital_status = df['marital'].unique()
print("marital_status with clean data:\n", unique_marital_status)

# 4. Filter the DataFrame for clients interested and not interested in the certificate of deposit
interested_clients = df[df['y'] == 'yes']
not_interested_clients = df[df['y'] == 'no']

# 5. Count the distribution of marital status among interested and not interested clients and print
interested_marital_distribution = interested_clients['marital'].value_counts()
not_interested_marital_distribution = not_interested_clients['marital'].value_counts()
print("Marital status distribution among clients interested in Certificate of Deposit:\n")
print(interested_marital_distribution)
print("Marital status distribution among clients not interested in Certificate of Deposit:\n")
print(not_interested_marital_distribution)
print('Number of surveys:\n' ,df.shape)

#6.  Plot the distribution of marital status among interested and not interested clients
fig, ax = plt.subplots()
interested_marital_distribution.plot(kind='bar', color='green', ax=ax, position=0, width=0.4, label='Interested')
not_interested_marital_distribution.plot(kind='bar', color='pink', ax=ax, position=1, width=0.4, label='Not Interested')
plt.title('Marital Status Distribution Among Clients')
plt.xlabel('Marital Status')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 7. To find all the unique values in a column of a DataFrame
unique_education_status = df['marital'].unique()
print('Status or civil marital with clean data:\n',unique_education_status)

# 8. Save changes to the Parquet file
df.to_parquet(parquet_file_path, index=False)

# 9. To find all the unique values in a column of a DataFrame
unique_education_status = df['marital'].unique()
print('Status or civil marital with clean data:\n',unique_education_status)
```

```python
# Made by Javier Coronado for the Talento Tech data analysis and visualization bootcamp, Mintic, Bogota, Colombia.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the Parquet file
parquet_file_path = "dataset_banco_1_2.parquet"
df = pd.read_parquet(parquet_file_path)

# Filter the dataframe for clients who answered 'yes' and 'no' in the 'y' column
yes_clients = df[df['y'] == 'yes']
no_clients = df[df['y'] == 'no']

# Group the data by the 'contact' column and count the occurrences for 'yes' and 'no'
yes_contact_counts = yes_clients['month'].value_counts()
no_contact_counts = no_clients['month'].value_counts()
print('\n', 'Number of clients who said yes:',yes_contact_counts, '\n')
print('\n','Number of clients who said no:', no_contact_counts, '\n')

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x=yes_contact_counts.index, y=yes_contact_counts.values, color='green', label='Yes')
sns.barplot(x=no_contact_counts.index, y=no_contact_counts.values, color='red', label='No', bottom=yes_contact_counts.values)
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Frequency of Contact by Conversion Status')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
