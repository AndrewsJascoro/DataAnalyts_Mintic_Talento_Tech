```python
# Ejemplo con datos nulos y duplicados usando pandas
# Por Javier Coronado para el curso Analisis de datos y visualizacion de Talento Tech, Mintic.
#"""2 de 4"""
import pandas as pd
from IPython.display import display
# Crear un DataFrame con altura, peso, edad y género
data1 = {'id': [1, 2, 3, 4, 5, 6, 7, 8],
         'weight': [144.3, 115.9, None, 122.6, 110.8, 100.89, 98.5, 115.9],
         'height': [7.5, 5.2, 7.5, 8.8, None, 5.9, 6.8, 7.5],
         'age': [25, 55, 41, 23, None, 38, 77, 41],
         'gender': ['F', 'M', 'F', 'F', 'F', 'M', 'M', 'F']}
df1 = pd.DataFrame(data1)
display('Mi Data Frame:', df1)

# Puedo agregar datos duplicados
data1_duplicate = {'id': [8, 9, 10],
                   'weight': [115.9, 98.5, 100.89],
                   'height': [7.5, 6.8, 5.9],
                   'age': [41, 77, 38],
                   'gender': ['F', 'M', 'M']}
# Creando así un nuevo DataFrame df1 que contiene tanto los datos originales como los datos duplicados
df1_duplicate = pd.DataFrame(data1_duplicate)
display('Datos Duplicados:' , df1_duplicate)

df1 = pd.concat([df1, df1_duplicate])
display('Datos concatenados dos DataFrames los df1 y df1_duplicate en forma vertical:' , df1)

# Mostrar información sobre los datos nulos y los duplicados
print("Información sobre datos nulos:")
display(df1.isnull().sum())

print("\nInformación sobre datos duplicados:")
display(df1.duplicated().sum())

# Mostrar el DataFrame resultante
print("\nDataFrame con datos nulos y duplicados:")
display(df1)
```

```python
# Ejemplo con datos nulos y duplicados usando pandas
# Por Javier Coronado para el curso Analisis de datos y visualizacion de Talento Tech, Mintic.
#"""4 de 4"""
import pandas as pd
from IPython.display import display

# Crear un DataFrame con altura, peso, edad y género
data1 = {'id': [1, 2, 3, 4, 5, 6, 7, 8],
         'weight': [144.3, 115.9, None, 122.6, 110.8, 100.89, 98.5, 115.9],
         'height': [7.5, 5.2, 7.5, 8.8, None, 5.9, 6.8, 7.5],
         'age': [25, 55, 41, 23, None, 38, 77, 41],
         'gender': ['F', 'M', 'F', 'F', 'F', 'M', 'M', 'F']}
df1 = pd.DataFrame(data1)
display("Mi Data Frame:", df1)

# Puedo agregar datos duplicados
data1_duplicate = {'id': [8, 9, 10],
                   'weight': [115.9, 98.5, 100.89],
                   'height': [7.5, 6.8, 5.9],
                   'age': [41, 77, 38],
                   'gender': ['F', 'M', 'M']}
# Creando así un nuevo DataFrame df1 que contiene tanto los datos originales como los datos duplicados
df1_duplicate = pd.DataFrame(data1_duplicate)
display("Datos Duplicados:", df1_duplicate)

# Concatenar los DataFrames para introducir datos duplicados
df1 = pd.concat([df1, df1_duplicate])
display("Datos concatenados dos DataFrames los df1 y df1_duplicate en forma vertical:", df1)

# Mostrar información sobre los datos nulos y los duplicados
display("Información sobre datos nulos:", df1.isnull().sum())
display("Información sobre datos duplicados:", df1.duplicated().sum())

# Filtrar los datos nulos y duplicados
df1_filtrado = df1.dropna()  # Eliminar filas con datos nulos
df1_filtrado = df1_filtrado.drop_duplicates()  # Eliminar filas duplicadas

# Mostrar el DataFrame resultante filtrado
display("DataFrame filtrado sin datos nulos y duplicados:", df1_filtrado)
```

```python
import pandas as pd
import matplotlib.pyplot as pl

# Define the path for the Parquet file
parquet_file_path = "dataset_banco_101_cleaned_copy.parquet" #"dataset_banco_101_cleaned_copy.parquet"

# Read the Parquet file into a pandas DataFrame
df = pd.read_parquet(parquet_file_path)

# To find all the unique values in a column of a DataFrame
unique_education_status = df['education'].unique()
print("This is unique objects in'education' column was found: \n",'\n', unique_education_status)

# Correct typographical errors in the 'education' column
corrections = {
    'SECONDARY': 'secondary',
    'Secondary': 'secondary',
    'UNK': 'unknown',
    'Primary': 'primary',
    'Tertiary': 'tertiary',
    'sec.': 'secondary'
}
df['education'] = df['education'].replace(corrections)

# Print the changes made
print("\n", "Typographical errors corrected:  \n")
for incorrect, correct in corrections.items():
    print(f"{incorrect} -> {correct}")

# To find all the unique values after in a column of a DataFrame
unique_education_status = df['education'].unique()
print("\n", "This is unique objects in'education'column was found after clean:\n", "\n", unique_education_status)

# Filter the DataFrame for clients interested and not interested in the certificate of deposit
interested_clients = df[df['y'] == 'yes']
not_interested_clients = df[df['y'] == 'no']

# Calculate the average age for interested and not interested clients
average_age_interested = interested_clients['age'].mean()
average_age_not_interested = not_interested_clients['age'].mean()

print("Average age of clients interested in Certificate of Deposit:", average_age_interested)
print("Average age of clients not interested in Certificate of Deposit:", average_age_not_interested)
```
