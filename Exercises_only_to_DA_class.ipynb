**The technical term for incorrect, inconsistent, or meaningless data in a dataset is "dirty data."**

**Note:**

**Data entry errors:** Typos, misspellings, and incorrect values entered by humans.

**Inconsistent data:** Inconsistent formatting, units, or representations of the same data across different records or columns.

**Missing data:** Empty or null values where data should be present.

**Duplicate data:** Duplicate records or entries within the dataset.

**Outliers:** Data points that are significantly different from other data points in the dataset, potentially indicating errors or anomalies.

**Abbreviations and acronyms:** Inconsistent or ambiguous abbreviations or acronyms used within the data.

**Inaccurate data:** Data that is outdated, incorrect, or no longer relevant.

**Non-standard values:** Values that do not conform to the expected range or format for a particular data field.

**Duplicate data:** Data that is poorly written, unclear, or unintelligible.

**Inconsistent capitalization:** Inconsistent use of uppercase and lowercase letters in text data.
