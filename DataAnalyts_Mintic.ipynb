```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Provided data
'''The provided block of code initializes a Python dictionary called data. 
This dictionary contains four key-value pairs:'''
data = {
    'Region': ['Central', 'Central', 'Central', 'Central', 'Central', 'Central', 'Este', 'Este', 'Este', 'Este', 'Este', 'Este', 'Oeste', 'Oeste', 'Oeste', 'Oeste', 'Oeste', 'Oeste'],
    'Producto': ['Accesorios', 'Accesorios', 'Dispositivos', 'Dispositivos', 'Sistemas', 'Sistemas', 'Accesorios', 'Accesorios', 'Dispositivos', 'Dispositivos', 'Sistemas', 'Sistemas', 'Accesorios', 'Accesorios', 'Dispositivos', 'Dispositivos', 'Sistemas', 'Sistemas'],
    'Vendedor': ['David', 'Karen', 'David', 'Karen', 'David', 'Karen', 'Ana', 'Lucas', 'Ana', 'Lucas', 'Ana', 'Lucas', 'Kevin', 'Sara', 'Kevin', 'Sara', 'Kevin', 'Sara'],
    'Ventas': [8287, 6909, 11420, 12948, 20098, 30633, 9323, 7667, 10348, 9312, 13531, 13374, 4744, 5442, 10711, 8780, 32855, 23151]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Ventas Data:\n", df)

# Check for missing values
# print("\nMissing values:\n", df.isnull().sum())

# Summary statistics
print("\n^Summary statistics:^\n", df.describe())

# Total sales per region
region_sales = df.groupby('Region')['Ventas'].sum().sort_values(ascending=False)
print("\nTotal Ventas per region:\n", region_sales)

# Total ventas per Vendedor
vendedor_sales = df.groupby('Vendedor')['Ventas'].sum().sort_values(ascending=False)
print("\nTotal Ventas per Vendedor:\n", vendedor_sales)

# Total ventas per Producto
producto_sales = df.groupby('Producto')['Ventas'].sum().sort_values(ascending=False)
print("\nTotal Ventas per Producto:\n", producto_sales)

# Grand Total Sales
grand_total_sales = df['Ventas'].sum()
print("\nGrand Total Ventas: $", grand_total_sales)

# Visualizing sales by Region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Ventas', data=df, estimator=sum, errorbar=None)
plt.title('Total Ventas by Region')
plt.show()
plt.savefig('sales_plot_region.png')

# Show me in a Plot Visualizing ventas by vendedor
plt.figure(figsize=(12, 6))
sns.barplot(x='Vendedor', y='Ventas', data=df, estimator=sum, errorbar=None)
plt.title('Total Ventas by Vendedor')
plt.show()

# Visualizing ventas by producto
plt.figure(figsize=(12, 6))
sns.barplot(x='Producto', y='Ventas', data=df, estimator=sum, errorbar=None)
plt.title('Total Ventas by Producto')
plt.show()

# Save the plot as an image file
plt.savefig('sales_plot_vendedor.png')

# Close the plot to avoid the warning
plt.close()
```
